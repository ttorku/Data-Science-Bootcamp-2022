```python
x1 =np.linspace(-3, 3, 100)

y1 =np.sin(x1)
y2 =np.cos(x1)

plt.plot(x1, y1, 'r')
plt.plot(x1, y2, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title("$y =\sin(x),  y=\cos(x)$")
plt.legend(['$\sin(x)$', '$\cos(x)$'], loc=4)
# plt.ylim(-0.75, 0.75)
# plt.xlim(-2, 2)
plt.savefig('first_graph.png')
plt.show()
```

### Creating Figures and Subplots

```python
fig=plt.figure(figsize=(12, 6))
ax =fig.add_axes([0, 0, 1,1])


ax.plot(x1, y1)
ax.plot(x1, y2)
```

```python
yy1 =3*x**2 +2*x -2
yy2 =4*x**2 +3
yy3 =-x**2 -5
yy4 =5*x**2 +3*x


fig1, ax =plt.subplots(2, 2)

ax[0,0].plot(x, yy1)
ax[0,0].set_xlabel('x')

ax[0, 1].plot(x, yy2)
ax[0,1].set_xlabel('x')

ax[1, 0].plot(x, yy3)
ax[1,0].set_xlabel('x')

ax[1,1].plot(x, yy4)
ax[1,1].set_xlabel('x')

# fig.subplots_adjust(left=0.5, bottom=1.0, right=0.9, top=0.2, wspace=0.8, hspace=0.2)
plt.show()
```

```python
arr1= np.array([5.7, 5.6, 6.5, 6.1, 4.5, 5.9, 5.8, 5.8, 5.6, 4.9]).reshape((-1,1))
arr2=np.array([50, 55, 56, 65, 75, 60, 80, 51, 52, 53]).reshape((-1,1))
arrdata =np.hstack((arr1, arr2))
```

```python
twodata =pd.DataFrame(arrdata, columns=["Height", "Weight"])
```

```python
twodata
```

```python
#merge the data
persondata['height'] =twodata['Height']
persondata['weight'] =twodata['Weight']
persondata.to_csv('person_update.csv')
```

```python
fig =plt.figure(figsize=(6,6))
axx =fig.add_axes([0, 0, 1, 1])
axx.plot(persondata['Age'], '--r')
axx.plot(persondata['height'], '-ob')
axx.plot(persondata['weight'], '-+g')
axx.legend(['Age', 'height', 'weight'])
axx.set_xlabel('People')
axx.set_ylabel('Values')
axx.set_title('Auto Data')
plt.savefig("person.png")
plt.show()
```
