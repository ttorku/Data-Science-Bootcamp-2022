Cleaning the Data of Publication column

*Rules to follow*
- Remove the extra dates in square brackets, wherever present: 1879 [1878]
- Convert date ranges to their “start date”, wherever present: 1860-63; 1839, 38-54
- Completely remove the dates we are not certain about and replace them with NumPy’s NaN: [1897?]
- Convert the string nan to NumPy’s NaN value

```python
book_new
```

```python
#1879 [1878]
def cleanDate(date):
  new_str = ""
  for char in date:
    if char != "[" and char != "," and char != "-" and char != "." \
    and char != "]" and char != "?" and char.isdigit() and char != "/":
      new_str += char
      # print(new_str)
    else:
      break
  new_str = new_str.strip()
  if len(new_str) < 4:
    new_str = np.nan
  return float(new_str)



sample_string = "1879 [1878]"
sample_string1 = "[1833]"
sample_string2 = "17 vols"
print(cleanDate(sample_string))
print(cleanDate(sample_string1))
print(cleanDate(sample_string2))

date_lst = []
for date in book_new["Date of Publication"]:
  date = str(date)
  cleaned_date = cleanDate(date)
  date_lst.append(cleaned_date)

book_new["Cleaned Date of Publication"] = date_lst
```

```python
from pandas.core import strings
import re
#search method

patt = "123"
string = "foo123bthey"

if re.search(patt, string):
  print("Search keyword found")
else:
  print("Search keyword does not exist")
```

```python
#metacharacters
#1.character class []

patt = "1"
string = "foo123bthey"

if re.search(patt, string):
  print("Search keyword found")
else:
  print("Search keyword does not exist")
```

```python
#match any number
patt = '[0-9]'
string = "foobthey"

if re.search(patt, string):
  print("Search keyword found")
else:
  print("Search keyword does not exist")
```

```python
#match three consecutive numbers
patt = '[0-9][0-9][0-9]'
string = "foobthey"

if re.search(patt, string):
  print("Search keyword found")
else:
  print("Search keyword does not exist")

string = "foob234they"

if re.search(patt, string):
  print("Search keyword found")
else:
  print("Search keyword does not exist")
```
